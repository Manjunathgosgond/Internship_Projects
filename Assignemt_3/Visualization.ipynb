```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import os
```

```python
os.listdir(input_path)
```

```python
Airbnb = pd.read_csv(input_path+"Airbnb_Dataset.csv")
Airbnb.head()
```

```python
Airbnb.shape
```

```python
Airbnb.isnull()
```

```python
Airbnb.isnull().sum()
```

```python
Airbnb.drop_duplicates()
```

```python
sns.countplot(data=Airbnb, x='neighbourhood_group', hue='room_type').set_title('Distribution of Room Types Across Different Neighbourhood Groups')
```

This count plot shows the distribution of different type of room in the neighbourhood group.

```python
sns.boxplot(data=Airbnb, x='neighbourhood_group', y='price').set_ylim(0,1000)
plt.title('Price distribution w.r.t neighbhourhood group')
```

This shows the price in differnet neighbhourhood group 


```python
sns.scatterplot(data=Airbnb, x='availability_365', y='price', hue='room_type').set_title('price vs avilability')
```

We can observe that the availability of the entire home/apt are more compare to the other two types of rooms and less available are the shared room. T he shared room value is not more than 200 in the available category.

```python
top_neighbourhoods = Airbnb['neighbourhood'].value_counts().head(10)
sns.barplot(y=top_neighbourhoods.index,x=top_neighbourhoods.values).set_title('top 10 neighbhourhoods listings')
```

From this we can conclude that witch neighbhourhood has more number of listing of rooms.

```python
avg_price_room_type = Airbnb.groupby('room_type')['price'].mean().reset_index()
sns.barplot(data=avg_price_room_type, x='room_type', y='price').set_title('avg_price_per_room_type')
```

We can get the information about the avg price of each room.

```python
HR = pd.read_csv(input_path+"HR_Dataset.csv")
HR.head()
```

```python
HR.shape
```

```python
HR.isnull().sum()
```

```python
HR.drop_duplicates
```

```python
HR= HR.drop(['MarriedID','MaritalStatusID','GenderID','EmpStatusID','DeptID','PerfScoreID','FromDiversityJobFairID'],axis=1)
```

```python
HR
```

```python
sns.histplot(x='Sex',data=HR).set_title('gender counts')
```

From this we can conclude that number of female employees are more than male employee.

```python
avg_salary_per_position=HR.groupby('Position')['Salary'].mean().reset_index()
print(avg_salary_per_position)
```

```python
sns.barplot(data=avg_salary_per_position,x='Salary',y='Position').set_title('avg salary per position')
```

We can get the information about the avg salary per position.

```python
count=HR.groupby('RecruitmentSource').sum().reset_index()
print(count)
```

```python
plt.pie(x=count.index,labels='RecruitmentSource',data=count,autopct='%1.1f%%')
```

This shows percentage distribution of recrutmentsource from where the more employees are hired.
